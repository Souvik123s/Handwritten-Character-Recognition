```python
import numpy as np
import os
import matplotlib.pyplot as plt
import scipy
import sklearn
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from skimage.transform import resize
import cv2
```

```python
#load the images from the dataset folder
def load_images_from_folder(folder):
    images = []
    for filename in os.listdir(folder):
        img = cv2.imread(os.path.join(folder,filename))
       # grayImage = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)  #convert each images to there greyscale imgae from there RGB
        ret, img = cv2.threshold(img,127,255,cv2.THRESH_BINARY) #convert each images to there binary imgae from there greyscale
        if img is not None:
            images.append(img)
    return images
```

```python
#load the images from the dataset folder
dataset=load_images_from_folder("C:/Users/LENOVO/project/CharacterDataset")
images=[]
for img in dataset:
    #convert the binaryimages to there 28*28 binary frame
    c=resize(img, (28, 28), mode='constant', cval=1.0, clip=True)
    c=c.flatten()
    images.append(c)
dataset=images
print("Database Contains ",len(dataset)," images")
#load the label dataset
df= pd.read_excel (r'C:\Users\LENOVO\project\CharacterDataset.xlsx')
labels=np.array(df)
labels=labels.astype(np.str)
```

```python
testing_ratio=[0.5,0.4,0.3,0.2]
for test in testing_ratio:
    print("for the ratio : ",test)
    accuracy=[]
    for it in range(0,20):
        #split the datasets randomly for test and train 
        X_train, X_test, Y_train, Y_test = train_test_split(dataset, labels, test_size=test, random_state=0)

        # Standardising the values
        Scaler = StandardScaler().fit(X_train)
        X_pca_train = Scaler.transform(X_train)

        # Call the PCA method with 100 components.
        pca = PCA(n_components=100)
        pca.fit(X_pca_train)
        P_train = pca.transform(X_pca_train)

        # Standardising the values
        X_pca_test = Scaler.transform(X_test)

        # Call the PCA method with 100 components.
        P_test = pca.transform(X_pca_test)

        # MLP Classifier with PCA (Principal Component Analisis) features
        mlp_classifier_PCA = MLPClassifier(hidden_layer_sizes=(300,400,150), max_iter=5000, tol=0.0001, random_state=1)

        # train the PCA model
        mlp_classifier_PCA.fit(P_train,Y_train.ravel())

        # test the model with tesing data
        mlp_prediction_PCA = mlp_classifier_PCA.predict(P_test)

        #print the accuracy of model
        count=0
        for ind in range(0,len(mlp_prediction_PCA)):
            if mlp_prediction_PCA[ind]==Y_test[ind]:
                count=count+1
        curr_accuracy=count*100/len(mlp_prediction_PCA);
        print("Iteration : ",it+1,"accuracy : ",curr_accuracy);
        accuracy.append(curr_accuracy)
    avg_accuracy=sum(accuracy)/20
    print("Avarage accuracy : ", avg_accuracy)
```
